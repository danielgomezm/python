```python
x = np.linspace(0, 1, 500)
y = np.sin(4 * np.pi * x) * np.exp(-5 * x)

plt.fill(x, y, zorder=10)
plt.grid(True, zorder=5)
```

```python
a = b = np.arange(0, 3, .02)
c = np.exp(a)
d = c[::-1]

plt.plot(a, c, 'k--', label='Model length')
plt.plot(a, d, 'k:', label='Data length')
plt.plot(a, c + d, 'k', label='Total message length')

legend = plt.legend(loc='upper center', shadow=True, fontsize='x-large')

legend.get_frame().set_facecolor('#ddddff')
```

```python
dt = 0.01
t = np.arange(0, 10, dt)
nse = np.random.randn(len(t))
r = np.exp(-t/0.05)

cnse = np.convolve(nse, r)*dt
cnse = cnse[:len(t)]
s = 0.1*np.sin(2*np.pi*t) + cnse

graph = plt.plot(t, s)
```

```python
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
label = plt.ylabel('Damped oscillation')
```

```python
x = np.linspace(0, 10)
plt.plot(x, np.sin(x), '--', linewidth=2)
plt.plot(x, np.cos(x), '--', linewidth=2)

fig = plt.gcf()
```

```python
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
plt.pie(sizes, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
axis = plt.axis('equal')
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

graph = plt.scatter(x, y, s=area, c=colors, alpha=0.5)
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
type(x)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
axe = ax2.scatter(x, y)
```

```python
f, axarr = plt.subplots(2, 2, subplot_kw=dict(projection='polar'))
axarr[0, 0].plot(x, y)
axarr[0, 1].scatter(x, y)
axarr[1, 0].plot(x, y ** 2)
axarr[1, 1].scatter(x, y ** 2)
f.subplots_adjust(hspace=0.3)
```
